```python
df['ratings'] = df['ratings'].astype(float)
```

#### CLEANING 'CUISINE' COLUMN

```python
df['cuisine']
```

```python
df['cuisine'] = df['cuisine'].str.split(',')
```

```python
df = df.explode('cuisine')
```

```python
df['cuisine'] = df['cuisine'].str.strip()
```

```python
food_map = {
    'Pizza': 'Pizza',
    'Pasta': 'Pasta',
    'Biryani': 'Biryani',
    'Momos': 'Momos',
    'Burger': 'Burger',
    'Fast Food': 'Fast Food',
    'Sandwich': 'Sandwich',
    'Wraps': 'Wraps',
    'Rolls': 'Rolls',
    'Desserts': 'Dessert',
    'Cake': 'Dessert',
    'Bakery': 'Dessert',
    'Mithai': 'Mithai',
    'Beverages': 'Beverages'
}

df['food_type'] = df['cuisine'].map(food_map).fillna('Main Course')
```

```python
cuisine_map = {
    'Pizza': 'Italian',
    'Pasta': 'Italian',
    'Burger': 'American',
    'Momos': 'Chinese',
    'Noodles': 'Chinese',
    'Sichuan': 'Chinese',
    'North Indian': 'North Indian',
    'South Indian': 'South Indian',
    'Andhra': 'Indian',
    'Biryani': 'Indian',
    'Kebab': 'Indian',
    'Sandwich': 'Fast Food',
    'Wraps':'Western',
    'Rolls':'Fast Food',
    'Salad':'Western',
    'Shawarma':'Middle Eastern',
    'Ice Cream':'Dessert',
    'Bakery': 'Dessert',
    'Shake': 'Beverages',
    'Pancake': 'Dessert',
    'Tea': 'Cafe',
    'Coffee':'Cafe',
    'Waffle': 'Dessert',
    'Juices':'Beverages',
    'Paan':'Dessert',
    'Frozen Yogurt':'Dessert',
    'Mithai':'Dessert'
}
```

```python
df['Cuisine_Main'] = df['cuisine'].map(cuisine_map)
```

```python
df['Cuisine_Main'] = df['Cuisine_Main'].fillna(df['cuisine'])
```

```python
df
```

```python
# Since one to many dataset is there we need to drop duplicates and create a new dataset

df_rest = df.drop_duplicates(subset=['names'])
```

```python
df_rest
```

## UNIVARIATE ANALYSIS

```python
import matplotlib.pyplot as plt
```

```python


plt.hist(df_rest['ratings'], bins=20)
plt.title("Distribution of Ratings")
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.show()
```

```python
df_rest['ratings'].mean()
```

#### INFERENCE
Ratings are mostly concentrated in 4-4.2.
Distribution is LEFT SKEWED (Mean < Median < Mode)
Average rating is 3.9 which indicated the overall customer satisfaction. 

#### Price for two people 

```python
plt.hist(df['price for one'] * 2, bins=20, edgecolor='black')
plt.title("Cost for Two Distribution")
plt.xlabel("Cost")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### INFERENCE
Most resturants will cost 200 for for people to eat in the resturant, the graph is RIGHT SKEWED. 

```python
df['Cuisine_Main'] = df['Cuisine_Main'].replace({'Desserts': 'Dessert'})
df['Cuisine_Main'].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))
df['Cuisine_Main'].value_counts().head(10).plot(kind='bar')
plt.xticks(rotation=45)
plt.show()
```

```python
df_rest.sort_values(by='ratings', ascending=False) \
    .head(15) \
    .plot(x='names', y='ratings', kind='barh', legend=False)

plt.title("Top 15 Highest Rated Restaurants")
plt.xlabel("Rating")
plt.ylabel("Restaurant")
plt.show()
```

# BIVARIATE ANALYSIS

```python
plt.scatter(df_rest['price for one'], df_rest['ratings'])
plt.title("Price vs Rating")
plt.xlabel("Price for One")
plt.ylabel("Rating")
plt.show()
```

```python
df_rest[['price for one', 'ratings']].corr()
```

#### INFERENCE 
There is no correlation between price and ratings

```python
df.groupby('Cuisine_Main')['ratings'] \
  .agg(['mean', 'count']) \
  .query('count > 30') \
  .sort_values(by='mean', ascending=False)
```

```python
df.groupby('Cuisine_Main')['price for one'] \
  .mean() \
  .sort_values(ascending=False)
```

```python
df.groupby('Cuisine_Main').agg({
    'ratings': 'mean',
    'price for one': 'mean'
}).sort_values(by='ratings', ascending=False)
```

```python
import seaborn as sns
sns.heatmap(df[['ratings','price for one']].corr(), annot=True)
```

# INSIGHTS 

1. There is a noticeable scope for improvement as most resturants are ranked 3.5-4 according to the data
2. Some cuisines offer high ratings at relatively lower prices, indicating strong value-for-money segments.
3. Certain cuisines consistently receive higher ratings than others, i.e., high-performing cuisines = safe investment areas
Low-performing cuisines = need improvement or repositioning
4. Most restaurants fall within a mid-price range, indicating a highly competitive segment.
